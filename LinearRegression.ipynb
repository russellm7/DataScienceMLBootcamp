```python
import pandas as pd
from pandas import DataFrame as df
from matplotlib import pyplot as pplot
```

```python
data=pd.read_csv("cost_revenue_clean.csv")
```

```python
data.describe()
```

```python
X=df(data,columns=["production_budget_usd"])
Y=df(data,columns=["worldwide_gross_usd"])
```

```python
pplot.figure(figsize=(10,6))
pplot.scatter(X,Y,alpha=0.3)
pplot.title("Film Cost vs Global Revenue")
pplot.xlabel("Production Budget ($)")
pplot.ylabel("Worldwide Gross ($)")
pplot.xlim(0,450000000)
pplot.ylim(0,3000000000)
pplot.show()
```
